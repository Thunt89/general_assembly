# Part 1: Regression

We will focus on **white wines only** for this part.

We will attempt to predict white wine (perceived) **quality** using information about the wine.

#### 1: Read in the white wine data and perform exploratory data analysis on it. Be sure to look for things like:

- missing values
- strange values
- relationships between features and the target (quality)

If you have trouble reading the file in, a good idea is to open it in a simple text editor to investigate its formatting!

```python
# Import data (note csv required text to column adjustment prior to load)
import pandas as pd
df = pd.read_csv('data/winequality-white.csv')
```

```python
# Check data types to see if any are not as expected - seems all are numerical, so fine
df.info()
```

```python
# Check for any null values - none found!
df.isnull().sum()
```

```python
# Check to see if there are any outlier values (min/max), or any features which are heavily skewed
df.describe()
```

```python
# Plot features which seem highly skewed to have a look - residual sugar, total sulfur dioxide
import matplotlib.pyplot as plt
%matplotlib inline

fig, axs = plt.subplots(1, 2, sharey=True)
axs[0].hist(df["residual sugar"], bins=30)
axs[0].legend(loc='upper right');
axs[1].hist(df["total sulfur dioxide"], bins=30);
axs[1].legend(loc='upper right');
```

```python
# Remove the outliers with high sugar contents (18 rows removed)
df.drop(df[df['residual sugar'] > 20].index, inplace=True)
# Remove the outliers with high total sulfur dioxide (24 rows removed)
df.drop(df[df['total sulfur dioxide'] > 250].index, inplace=True)
```

```python
# Plot correlation map to understand which features may influence the target
import seaborn as sns
import numpy as np
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask,vmin=-1, vmax=1, square=True, cmap="coolwarm")
```

```python
# Note some of the values are quite close, so have a look at the numbers with: df.corr() 
```

Perceived Quality appears to be significantly correlated with: alcohol, density(negatively), chlorides(negatively) 

#### 2: Do a train-test split and choose 3 predictors to predict quality

Remember: we want to avoid testing on the final test set until the end.

```python
from sklearn.model_selection import train_test_split
X = df[["alcohol","density", "chlorides"]]
y = df["quality"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(len(X_train), len(X_test))
print(len(y_train), len(y_test))
```

```python
df.head()
```

```python
# Plot the features on a scatter plot to have a look at the distributions and relationships with quality
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3,figsize=(20,6), sharey=True, sharex=False )
sns.regplot(X_train['alcohol'], y_train, ax=ax1, label="alcohol", scatter_kws={"s": 100})
sns.regplot(X_train['chlorides'], y_train, ax=ax2, label="chlorides", scatter_kws={"s": 100})
sns.regplot(X_train['density'], y_train, ax=ax3, label="density", scatter_kws={"s": 100})
ax1.set_ylim(top=10, bottom=2) 
ax3.set_xlim([0.985, 1.005]) 
plt.show()
```

```python
# Use pairplot to look at how predictor/target distributions might correlate with each other
predictors = ["alcohol", "chlorides", "density"]
target = ["quality"]
sns.pairplot(df[predictors + target])
plt.show()
```

#### 3: Now do *another* train-test split using the training set above.

We do this to simulate a train-test split so we can evaluate a model, **without** touching the test set we created in step 2.

We often call this second test set (the one it's OK to test on) the "validation" set.

```python
X2 = X_train[["alcohol","density", "chlorides"]]
y2 = y_train

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.3, random_state=42)

print(len(X_train2), len(X_test2))
print(len(y_train2), len(y_test2))
```

#### 3: Fit a linear regression on the training set and examine your interecept and coefficients. What do they tell you?

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train2, y_train2)
y_train_pred = model.predict(X_train.iloc[y_test2])
```

```python
# Training model interpretation
print(model.intercept_, model.coef_)
```

Model suggests that for every 1% increase in alcohol content, rating improves by 0.4. For every 1 unit increase in density the quality improves by 37 points (a nonsense) and quality decreases by 2.2 for every unit increase of chlorides

#### 4: How well does your model do? Calculate the RMSE on the validation set.

```python
from sklearn.metrics import mean_squared_error
rmse1 = np.sqrt(mean_squared_error(y_train.iloc[y_test2], y_train_pred))
print(f"\tRMSE: {rmse1}")
```

#### 5: Obtain the average RMSE score across 10 folds of cross-validation. This should give you a better idea of your model's performance on new data. How does it compare to the RMSE calculated above?

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

model = LinearRegression()

cv_scores = cross_val_score(model, X_train, y_train, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-cv_scores)
print(rmse_scores)
print("Mean RMSE: {0:.1f}".format(np.mean(rmse_scores)))
```

Model achieves better accuracy using 10 folds of cross-validation

#### 6: Now try using *all the features* with both a lasso and ridge regression. Obtain the best alpha values using grid search.

```python
# Using all features in the original dataframe, create a training and test data set
from sklearn.linear_model import LinearRegression

X = df.iloc[:,:-1]
y = df.iloc[:,-1:]

X_train3, X_test3, y_train3, y_test3 = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit a linear regression model
lr = LinearRegression()
lr.fit(X_train3, y_train3);
```

```python
# Fit a Ridge regularisation model (arbitrarily alpha = 100 to start)
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=100)
ridge.fit(X_train3, y_train3)

print("LinearRegression with all features RMSE:", np.sqrt(mean_squared_error(y_train3, lr.predict(X_train3))))
print("Ridge LinearRegression with all features RMSE:", np.sqrt(mean_squared_error(y_train3, ridge.predict(X_train3))))
```

```python
# Fit a Lasso  regularisation model
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(X_train3, y_train3)

print("LinearRegression with all features RMSE:", np.sqrt(mean_squared_error(y_train3, lr.predict(X_train3))))
print("Lasso LinearRegression with all features RMSE:", np.sqrt(mean_squared_error(y_train3, lasso.predict(X_train3))))
```

```python
# Do a grid search for best alpha value of ridge regularisation
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=Ridge(),
                    param_grid={'alpha': np.logspace(-10, 10, 21)},
                    scoring='neg_mean_squared_error',
                    return_train_score=True)

grid.fit(X_train3,y_train3)
```

```python
# How good is the Ridge LinearRegression model with optimal alpha value and when applied to the localised test data
print(f"Ridge LinearRegression with all features, RMSE:", np.sqrt(-grid.best_score_),"alpha:", grid.best_params_)
best_model = grid.best_estimator_
print(f"Model applied to local test set, RMSE:",np.sqrt(mean_squared_error(y_train3, best_model.predict(X_train3))))
```

#### 7: What are your conclusions based on the models you've fit above?

```python
print(f"LinearRegression, 3 features, 1 manual train-test-split, RMSE: {rmse1}")
print("LinearRegression, 3 features, cv=10, RMSE",(np.mean(rmse_scores)))
print("LinearRegression, 11 features, no cv, RMSE:", np.sqrt(mean_squared_error(y_train3, lr.predict(X_train3))))
print("Ridge (a=100) LinearRegression, 11 features, no cv, RMSE:", np.sqrt(mean_squared_error(y_train3, ridge.predict(X_train3))))
print("Lasso LinearRegression, 11 features, no cv, RMSE:", np.sqrt(mean_squared_error(y_train3, lasso.predict(X_train3))))
print(f"Ridge (a=1e-5), 11 features, no cv , RMSE:",np.sqrt(mean_squared_error(y_train3, best_model.predict(X_train3))))
```

Based on the 6 models run, the one with the least error appears to be the linear regression model with Ridge regularisation and alpha = 0.00001. This is only marginally better than using LinearRegression without Ridge - they have the same RMSE to 4d.p

### 8: Bonus - compare your findings to another regression model

Use the same 3 features you chose above, and try one of: KNN, a decision tree, a random forest, or something completely different!

**8.1:** first, get the cross-validated RMSE of your other model

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train2, y_train2)

y_KNN_pred = model.predict(X_train.iloc[y_test2])

rmse2 = np.sqrt(mean_squared_error(y_train.iloc[y_test2], y_KNN_pred))
print(f"KNeighborsRegressor, 3 features, K=3, 1 manual train-test-split, RMSE: {rmse2}")
```

```python
# Do a grid search to find the best value of K
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=KNeighborsRegressor(),
                    param_grid={"n_neighbors": [3, 5, 7, 9]},
                    scoring='neg_mean_squared_error',
                    return_train_score=True)

grid.fit(X_train2, y_train2)
```

```python
# How good is the KNN model with optimal K value and when applied to the localised test data
print(f"KNN, 3 features, K=9 RMSE:", np.sqrt(-grid.best_score_),"K:", grid.best_params_)
best_model2 = grid.best_estimator_
print(f"KNN, 3 features, K=9, on local test set, RMSE:",np.sqrt(mean_squared_error(y_test2, best_model2.predict(X_train.iloc[y_test2]))))
```

RMSE seems quite good, but then when KNN model is applied to local test data it performs quite poorly?!
