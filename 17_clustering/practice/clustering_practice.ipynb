### 4: Look at the distribution of your chosen variables. How close are they in scale?

```python
import matplotlib.pyplot as plt
def show_cluster_boxplots(df, columns, cluster_column):
    fig, ax = plt.subplots(1, len(columns), figsize=(16, 8))

    for i, c in enumerate(columns):
        sns.boxplot(x=cluster_column, y=c, data=df, ax=ax[i])
        ax[i].set_title(f"Box plot of {c} by cluster")
    
    plt.show()
```

```python
# Variables are quite far apart on scale, so standardisation is probably required
show_cluster_boxplots(beers, columns=["Efficiency", "BoilTime", "Color", "BoilGravity"], cluster_column="BrewMethod")
```

### 5: Scale your variables in preparation for clustering

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(beers[["Efficiency", "BoilTime", "Color", "BoilGravity"]])
```

### 6: Try k-means clustering by choosing a value of k (just a best guess for now!)

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)

clusters = kmeans.fit_predict(X)
clusters
```

### 7: Examine your results:

- what is the distribution (value counts) of your new labels?
- replot the pairplot coloured by your clusters - what do you think?

```python
# There is a majority class '0' and then a medium class '1' and a small class '2'
beers["cluster"] = clusters
beers["cluster"].value_counts()
```

```python
sns.pairplot(beers, hue="cluster", vars=["Efficiency", "BoilTime", "Color", "BoilGravity"]);
```

### 9: Use "inertia" and the "elbow method" to find a "better" value for k

```python
def plot_inertia_values(k_min, k_max):
    inertia_values = []
    k_values = list(range(k_min, k_max+1))

    for k in k_values:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        inertia_values.append(kmeans.inertia_)

    fig, ax = plt.subplots()
    ax.plot(k_values, inertia_values)
    ax.set_title("Inertia values for different $k$")
    ax.set_xlabel("k")
    ax.set_ylabel("Inertia")
    #ax.set_ylim((0, 300))
    ax.set_xticks(k_values)
    plt.show()
    
    return inertia_values, k_values
```

```python
iv, kv = plot_inertia_values(k_min=2, k_max=15)
```

### 10: choose an "optimal" value of $k$ that's not too large (less than 10-15). Refit your clusters if you didn't save the cluster assignments from step 9 (i.e. store which of your data points are assigned to which of your k clusters).

```python
kmeans = KMeans(n_clusters=5, random_state=42)

clusters = kmeans.fit_predict(X)
# Output result to DataFrame
beers["cluster"] = clusters
beers["cluster"].value_counts()
```
