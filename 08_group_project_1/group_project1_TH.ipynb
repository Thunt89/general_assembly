```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/video_game_sales.csv')   
df.head()
```

```python
df.dropna
```

```python
df.shape
```

```python
df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Genre')[['NA_Sales','EU_Sales']].sum().sort_values(by='NA_Sales',ascending=False)
```

```python
Global_sales_by_year = df.groupby('Year')['Global_Sales'].sum().sort_values(ascending=False)
Global_sales_by_year
```

```python
df.loc[:,['Year','Platform','Global_Sales']].groupby('Platform')['Global_Sales'].sum().T
```

```python
df.groupby(['Year']).agg({'NA_Sales':'sum','EU_Sales':'sum', 'JP_Sales':'sum', 'Other_Sales':'sum', 'Global_Sales':'sum'})
```

```python
platform_list = df['Platform'].unique()
item_list = []

fig, ax = plt.subplots(len(platform_list), 1)

for idx, item in enumerate(platform_list):
    item_list.append(item)
    item_list = []

    df[df['Platform'].isin(platform_list)].groupby('Year').agg({'Global_Sales':'sum'}).plot(ax= ax[idx])
```

```python
platform_list = df['Platform'].unique()
item_list = []

fig, ax = plt.subplots(figsize=(8,8))

for item in platform_list:
    item_list.append(item)

    df[df['Platform'].isin([item])].groupby('Year')['Global_Sales'].sum().plot(ax=ax, label=item)

ax.legend()
    
```

```python
df_stack = df.groupby(['Platform', 'Year'])['Global_Sales'].sum().unstack().T
df_stack
```

```python
df_stack.plot()
```

```python
top_list = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)[:10]

df2 = df[df['Publisher'].isin(top_list.index)]
df2

df_stack = df2.groupby(['Publisher', 'Genre'])['Global_Sales'].sum().sort_values(ascending=False).unstack().T
df_stack
df_stack.plot(kind='bar',figsize=(20,12))
```

```python
top_list = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)[:10]

df2 = df[df['Publisher'].isin(top_list.index)]
df2

df_stack = df2.groupby(['Publisher', 'Platform'])['Global_Sales'].sum().sort_values(ascending=False).unstack().T
df_stack
df_stack.plot(kind='bar',figsize=(20,12))

```
